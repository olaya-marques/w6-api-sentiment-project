## 3. WordCloud 

- Ejemplo de Finlandia:

*Pa铆s m谩s feliz seg煤n el 铆ndice global de la felicidad*

```python
def wordcloud(df,row,col):
    
    wordcloud=WordCloud(width=1600,height=400).generate(df.loc[row][col])
    plt.figure(figsize=(15, 10), facecolor='k')
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.tight_layout(pad=0)
    plt.show();
```

```python
wordcloud(data, 1 , "Speech")
```

- Dataset entero:

```python
def wordcloud_df(df, col):
    
    wordcloud=WordCloud(width=1600,
                        height=400).generate(' '.join([e for e in df[col]]))
    
    plt.figure(figsize=(15, 10), facecolor='k')
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.tight_layout(pad=0)
    plt.show();
```

```python
wordcloud_df(data, "Speech")
```
